```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import time
import datetime
import pandas as pd
#####################################
### Parameters to be changed here ###
#####################################
TimeSteps = 10
ticker = "MSFT"
# Route for saved model
sURL = "/content/drive/MyDrive/public/DREAM/"
filename = 'MSFT1643759940.csv'
resultsDF = pd.read_csv(sURL+filename)
resultsDF.Date = pd.to_datetime(resultsDF['Date'],format='%Y/%m/%d')
resultsDF.set_index('Date')

# other params
lastX = int(resultsDF.dif.size) - 250  # just visualize last year
res = resultsDF.dif[lastX:]
threshold = res.mean()      # set threshold = mean of last year (250 days)
print('mean =' + str(res.mean()))
```

```python
# Let's check the table of results
TP = 0   # true positive
FP = 0   # False positive
TN = 0   # True negative
FN = 0   # False negative
accurateTP = 0 # accurate True Positive
accurateTN = 0 # accurate True Negative
totalRes = resultsDF.realDif.size

for i, values in resultsDF.iterrows():
  if (resultsDF.realDif[i]>0 and resultsDF.predDif[i]>0):
    TP += 1
    if (resultsDF.realDif[i] - resultsDF.predDif[i]) < threshold:
        accurateTP +=1
  if (resultsDF.realDif[i] < 0 and resultsDF.predDif[i] > 0):
    FP += 1
  if (resultsDF.realDif[i] < 0 and resultsDF.predDif[i] < 0):
    TN += 1
    if (resultsDF.realDif[i] - resultsDF.predDif[i]) < threshold:
        accurateTN +=1
  if (resultsDF.realDif[i] > 0 and resultsDF.predDif[i] < 0):
    FN += 1
  i = i + 1

pTP = (TP/totalRes) * 100
pAccurateTP = (accurateTP/totalRes) *100
pFP = (FP/totalRes) * 100
pTN = (TN/totalRes) * 100
pAccurateTN = (accurateTN/totalRes) *100
pFN = (FN/totalRes) * 100

print("TotalRes =" + str(totalRes))
print("Total = " +str(TP+FP+TN+FN))
print("\nTP = " + str(TP) + "( %2f" % pTP + " )")
print("FP = " + str(FP) + "( %2f" % pFP + " )")
print("TN = " + str(TN) + "( %2f" % pTN + " )")
print("FN = " + str(FN) + "( %2f" % pFP + " )")

print("\nright guess = " + str(TP + TN) + "( %2f" % (pTP+pTN) + " )")
print("error = " + str(FP + FN) + "( %2f" % (pFP+pFN) + " )")
print("\nAccurate TP = " + str(accurateTP) + "( %2f" % pAccurateTP + " )")
print("Accurate TN = " + str(accurateTN) + "( %2f" % pAccurateTN + " )")
```
